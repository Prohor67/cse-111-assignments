```python
#TASK01
class Student:
  ID = 0
  def __init__(self,name,dept,age,cgpa):
    self.name = name
    self.dept = dept
    self.age = age
    self.cgpa = cgpa
    Student.ID+=1
    self.Id = Student.ID
  def showDetails(self):
    print(f"ID: {self.Id}\nName: {self.name}\nDepartment: {self.dept}\nAge: {self.age}\nCGPA: {self.cgpa}")

  @classmethod
  def from_String(cls,info):
    info_list = info.split("-")
    return Student(info_list[0],info_list[1],info_list[2],info_list[3])

s1 = Student("Samin", "CSE", 21, 3.91)
s1.showDetails()
print("-----------------------")
s2 = Student("Fahim", "ECE", 21, 3.85)
s2.showDetails()
print("-----------------------")
s3 = Student("Tahura", "EEE", 22, 3.01)
s3.showDetails()
print("-----------------------")
s4 = Student.from_String("Sumaiya-BBA-23-3.96")
s4.showDetails()


print("============================================")
print("SUBTASK - 6")
print("Class variable is the variable of the whole class, Instance variable is the variable of a specific object")
print("============================================")
print("SUBTASK - 7")
print("Instace method is accessible by only specific object but class method is accessible by all the method")
```

```python
#TASK02
class Assassin:
  num = 0
  def __init__(self,name,rate):
    self.name = name
    self.rate = rate
    Assassin.num += 1
    self.num = Assassin.num
  def printDetails(self):
    print(f"Name: {self.name}\nSuccess Rate: {self.rate}%\nTotal Numbers of Assassin: {self.num}")
  @classmethod
  def failureRate(cls,name,frate):
    return Assassin(name,100-frate)
  @classmethod
  def failurePercentage(cls,name,percantage):
    percantage = int(percantage.replace("%",""))
    return Assassin(name,100-percantage)

john_wick = Assassin('John Wick', 100)
john_wick.printDetails()
print('================================')
nagisa = Assassin.failureRate("Nagisa", 20)
nagisa.printDetails()
print('================================')
akabane = Assassin.failurePercentage("Akabane", "10%")
akabane.printDetails()


```

```python
#TASK07
class Cat:
  Number_of_cats = 0
  def __init__(self,color,work):
    self.color = color
    self.work = work
    Cat.Number_of_cats += 1
  @classmethod
  def no_parameter(cls):
    return Cat("White","sitting")
  @classmethod
  def first_parameter(cls,color):
    return Cat(color,"sitting")
  @classmethod
  def second_parameter(cls,work):
    return Cat("Grey",work)
  def changeColor(self,new):
    self.color = new
  def printCat(self):
    print(f"{self.color} is {self.work}")


print("Total number of cats:", Cat.Number_of_cats)
c1 = Cat.no_parameter()
c2 = Cat.first_parameter("Black")
c3 = Cat("Brown", "jumping")
c4 = Cat("Red", "purring")
c5 = Cat.second_parameter("playing")
print("=======================")
c1.printCat()
c2.printCat()
c3.printCat()
c4.printCat()
c5.printCat()
c1.changeColor("Blue")
c3.changeColor("Purple")
c1.printCat()
c3.printCat()
print("=======================")
print("Total number of cats:", Cat.Number_of_cats)
```

```python
#TASK08
import math
class Cylinder:
  radius = 5
  height = 18
  def __init__(self,radius,height):
    print(f"Default radius= {Cylinder.radius} and height = {Cylinder.height}.")
    Cylinder.radius = radius
    Cylinder.height = height
    self.radius = radius
    self.height = height
    print(f"Updated radius= {Cylinder.radius} and height= {Cylinder.height}.")

  @staticmethod
  def area(radius,height):
    print(f"Area : {float((2*math.pi*radius**2) + (2*math.pi*radius*height))}")

  @staticmethod
  def volume(radius,height):
    print(f"Volume : {float(math.pi*radius**2*height)}")

  @classmethod
  def swap(cls,height,radius):
    return Cylinder(radius,height)
  @classmethod
  def changeFormat(cls,string):
    data_list = string.split("-")
    return Cylinder(float(data_list[0]),float(data_list[1]))








c1 = Cylinder(0,0)
Cylinder.area(c1.radius,c1.height)
Cylinder.volume(c1.radius,c1.height)
print("===============================")
c2 = Cylinder.swap(8,3)
c2.area(c2.radius,c2.height)
c2.volume(c2.radius,c2.height)
print("===============================")
c3 = Cylinder.changeFormat("7-13")
c3.area(c3.radius,c3.height)
c3.volume(c3.radius,c3.height)
print("===============================")
Cylinder(0.3,5.56).area(Cylinder.radius,Cylinder.height)
print("===============================")
Cylinder(3,5).volume(Cylinder.radius,Cylinder.height)
```

```python
#TASK10
class SultansDine:
  Total_branch = 0
  Total_sell = 0
  branch_list = []
  def __init__(self,branch):
    self.branch = branch
    self.sell = 0
    self.branch_dic = {}
    SultansDine.Total_branch += 1
  def sellQuantity(self,sell):
    if sell<10:
      self.sell = sell*300
    elif sell<20:
      self.sell = sell*350
    else:
      self.sell = sell*400
    SultansDine.Total_sell += self.sell
    self.branch_dic[self.branch] = self.sell
    SultansDine.branch_list.append(self.branch_dic)

  def branchInformation(self):
    print(f"Branch Name: {self.branch}\nBranch Sell: {self.sell}")

  @classmethod
  def details(cls):
    print(f"Total Number of branch(s): {cls.Total_branch}")
    print(f"Total Sell: {cls.Total_sell} Taka")
    for i in cls.branch_list:
      for j in i:
        print(f"Branch Name: {j}, Branch Sell: {i[j]} Taka\nBranch consists of total sell's: {round((i[j]/cls.Total_sell)*100,2)}%")



SultansDine.details()
print('########################')
dhanmondi = SultansDine('Dhanmondi')
dhanmondi.sellQuantity(25)
dhanmondi.branchInformation()
print('-----------------------------------------')
SultansDine.details()

print('========================')

baily_road = SultansDine('Baily Road')
baily_road.sellQuantity(15)
baily_road.branchInformation()
print('-----------------------------------------')
SultansDine.details()

print('========================')

gulshan = SultansDine('Gulshan')
gulshan.sellQuantity(9)
gulshan.branchInformation()
print('-----------------------------------------')
SultansDine.details()
```
