```python
#TASK01
import math
class Circle:
  def __init__(self, radius = None):
    self.__radius = radius
  def getRadius(self):
    return self.__radius
  def setRadius(self,new):
    self.__radius = new
  def area(self):
    return math.pi*self.__radius**2





c1 = Circle(4)
print("First circle radius:" , c1.getRadius())
print("First circle area:" , c1.area())


c2 = Circle(5)
print("Second circle radius:" , c2.getRadius())
print("Second circle area:" , c2.area())
```

```python
#TASK02
class Triangle:
  def __init__(self,base  = None, height =None):
    self.__base = base
    self.__height = height
  def getBase(self):
    return self.__base
  def getHeight(self):
    return self.__height
  def setBase(self,new):
    self.__base = new
  def setHeight(self,new):
    self.__height = new
  def area(self):
    return 0.5*self.__base*self.__height

t1 = Triangle(10, 5)
print("First Triangle Base:" , t1.getBase())
print("First Triangle Height:" , t1.getHeight())
print("First Triangle area:" ,t1.area())


t2 = Triangle(5, 3)
print("Second Triangle Base:" , t2.getBase())
print("Second Triangle Height:" , t2.getHeight())
print("Second Triangle area:" ,t2.area())
```

```python
#TASK05
class Student:
  def __init__(self,name,id,dept,email = None, password = None, login_status = None):
    self.name = name
    self.id = id
    self.dept = dept
    self.email = email
    self.password = password
    self.login_status = login_status
    print("Student object is created!")
class Usis:
  def __init__(self):
    print("USIS is ready to use!")
  def login(self,student):
    if student.email == None and student.password == None:
      print("Email and password need to be set")
    else:
      print("Login successful!")
      student.login_status = "OK"
  def advising(self,student,*args):
    self.courses = args
    if student.login_status == None:
      print("Please login to advise courses!")
    else:
      if len(self.courses) == 0:
        print("You haven't selected any courses.")
      elif len(self.courses)>3:
        print("You need special approval to take more than 3 courses.")
      else:
        print("Advising successful!")
  def individualDetails(self,student):
    sub = ""
    for i in range(len(self.courses)):
      if i == len(self.courses)-1:
        sub += self.courses[i]
      else:
        sub += self.courses[i]+","+" "
    return f"Name: {student.name}\nID: {student.id}\nDepartment: {student.dept}\nAdvised courses: {sub}"






rakib = Student("Rakib", 12301455, "CSE")
print("1***********************")
usis_obj = Usis()
print("2***********************")
usis_obj.login(rakib)
print("3***********************")
usis_obj.advising(rakib)
print("4***********************")
rakib.email = "rakib@hotmail.com"
rakib.password = "1234"
print("5***********************")
usis_obj.login(rakib)
print("6***********************")
usis_obj.advising(rakib)
print("7***********************")
usis_obj.advising(rakib, "CSE110", "PHY111", "MAT110", "CSE260")
print("8***********************")
usis_obj.advising(rakib, "CSE110", "PHY111", "MAT110")
print("9***********************")
print(usis_obj.individualDetails(rakib))
```

```python
#TASK06
class Train:
  def __init__(self,name, *args):
    self.name = name
    self.station_list = args
    self.stations = {}
    self.passenger = []
    for i in range(len(args)):
      self.stations[args[i]] = i+1
    print(f"Welcome aboard on {self.name}")
    print(f"Start: {args[0]}")
    print(f"Destination: {args[-1]}")
  def addPassenger(self,*args):
    for i in args:
      self.passenger.append(i)
      print(f"{i.name} welcome aboard")

  def allPassengerDetails(self):
    for i in self.passenger:
      if i.destination == None:
        destination = self.station_list[-1]
      else:
        destination = i.destination
      if i.start == None:
        start = self.station_list[0]
      else:
        start = i.start
      fare = abs(self.stations[destination]-self.stations[start])*100
      print(f"Name: {i.name},Start: {start},Destination:\n{destination},Fare: {fare}$")





class Passenger:
  def __init__(self,name, start =None , destination = None):
    self.name = name
    self.start = start
    self.destination = destination








t1 = Train('T1-Express','New York','Manhattan','Brooklyn','Boston')
print("1========================")
p1 =Passenger("Naruto")
t1.addPassenger(p1)
p2 = Passenger("Sasuke","Manhattan")
p3 = Passenger("Hinata","Manhattan","Brooklyn")
print("2========================")
t1.addPassenger(p2,p3)
print("3========================")
t1.allPassengerDetails()
print("4========================")
t2 = Train('Europe-Express','London','Paris','Brussels','Turkey')
print("5========================")
p4 =Passenger("Max","London","Brussels")
p5 = Passenger("Eleven","Paris")
p6 = Passenger("Mike","Brussels")
t2.addPassenger(p4,p5,p6)
print("6========================")
t2.allPassengerDetails()
```

```python
#TASK07
class BracuStudent:
  def __init__(self,name,home):
    self.name = name
    self.home = home
    self.status = False
  def show_details(self):
    print(f"Student Name: {self.name}\nLives in {self.home}\nHave Bus Pass? {self.status}")
  def get_pass(self):
    self.status = True

class BracuBus:
  def __init__(self,place, seat =2):
    self.place = place
    self.seat = seat
    self.passengers = []
  def board(self,*args):
    if len(args) == 0:
      print("No passenger!")
    else:
      for i in args:
        if i.home != self.place:
          print("You got on wrong bus!")
        elif i.status == False:
          print("You don't have bus pass!")
        elif len(self.passengers)==self.seat:
          print("Bus is full!")
        else:
          self.passengers.append(i.name)
          print(f"{i.name} boarded the bus")
  def show_details(self):
    print(f"Bus Route: {self.place}\nPassengers Count: {len(self.passengers)} (Max: {self.seat})\nPassengers On Board: {self.passengers}")


st1 = BracuStudent("Afif", "Mirpur")
print("1===========================")
st2 = BracuStudent("Shanto", "Motijheel")
st3 = BracuStudent("Taskin", "Mirpur")
st1.show_details()
st2.show_details()
print("2===========================")
st3.show_details()
print("3===========================")
bus1 = BracuBus("Mirpur")
bus2 = BracuBus("Azimpur", 5)
bus1.show_details()
bus2.show_details()
print("4===========================")
st2.get_pass()
st3.get_pass()
print("5===========================")
st2.show_details()
st3.show_details()
print("6===========================")
bus1.board()
print("7===========================")
bus1.board(st1, st2)
print("8===========================")
st1.get_pass()
st2.home = "Mirpur"
st1.show_details()
st2.show_details()
print("9===========================")
bus1.board(st1, st2, st3)
print("10===========================")
bus1.show_details()
```

```python
#TASK08
class Library:
  def __init__(self,name,books):
    self.name = name
    self.books = books
    self.borrower = {}
  def details(self):
    print(f"{self.name} Library details\nBorrower details:\n{self.borrower}\nBooks availability:\n{self.books}")

class Reader:
  def __init__(self,name):
    self.name = name

    self.borrow_list = {}
    self.book_list = []
  def borrow(self,library,*args):
    self.library = library

    for i in args:
      if len(self.book_list)<5 and self.library.books[i] != 0:
        self.book_list.append(i)
        self.library.books[i] = self.library.books[i]-1
        if i in self.borrow_list:
          self.borrow_list[i] += 1
        else:
          self.borrow_list[i] = 1

        self.library.borrower[self.name] = len(self.book_list)
        print(f"{i} books borrowed successfully.")
      elif self.library.books[i] == 0:
        print(f"{i} books are not available at the moment.")
      elif len(self.book_list) == 5:
        print("You cannot borrow more than 5 books.")
  def readerInfo(self,book = None):
    if book == None:
      print(f"{self.name}, you have {len(self.book_list)} book(s) with you")
      for i in self.borrow_list:
        print(f"Books on {i}: {self.borrow_list[i]}")
    else:
      print(f"{self.name}, you have {self.borrow_list[book]} {book} book(s) with you.")

L1=Library('Dhaka',{'Arts':15,'Fiction':135,'Politics':2,'Science':11,'Poetry':15})
L1.details()
print("1----------------------")
r1=Reader('Aladdin')
r1.borrow(L1,'Arts','Fiction','Fiction','Politics')
print("2----------------------")
r1.borrow(L1,'Politics','Fiction')
print("3----------------------")
r1.readerInfo()
print("4----------------------")
r1.readerInfo('Fiction')
print("5----------------------")
L1.details()
print("6----------------------")
r2=Reader('Jasmine')
r2.borrow(L1,'Politics','Poetry')
print("7----------------------")
r2.readerInfo()
print("8----------------------")
L1.details()
```
